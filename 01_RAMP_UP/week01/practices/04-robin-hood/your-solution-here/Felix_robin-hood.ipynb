# Robin Hood: 
Estamos en plena competición para ganar el concurso de tiro con arco en Sherwood. Con nuestro arco y flechas disparamos sobre una diana e intentamos acertar lo más cerca posible del centro. 

El centro de la diana viene representado por los valores (0, 0) en los ejes de coordenadas.

## Descripción: 
En el espacio de 2 dimensiones un punto se puede definir por un par de valores que corresponden a la coordenada horizontal (x) y a la vertical (y). El espacio puede quedar dividido en 4 zonas (cuadrantes): Q1, Q2, Q3, Q4. Cuyo punto de unión único es el punto (0, 0). 

Si un punto se encuentra en Q1 tanto su coordenada x como la y son positivas. Te dejo un enlace a wikipedia para que te familiarices con estos cuadrantes.

```python
puntos = [(4,5), (-0,2), (4,7), (1,-3), (3,-2), (4,5), 
          (3,2), (5,7), (-5,7), (2,2), (-4,5), (0,-2), 
          (-4,7), (-1,3), (-3,2), (-4,-5), (-3,2), 
          (5,7), (5,7), (2,2), (9, 9), (-8, -9)]

# 1. Robin Hood es famoso por acertar a una flecha con otra flecha. ¿Lo ha conseguido? 

# 2. Calcula cuántos flechazos han caido en cada cuadrante. 

# 3. Halla el punto más cercano al centro. Calcula su distancia al centro
# Definir una función que calcula la distancia al centro puede servir de ayuda. 

# 4. Si la diana tiene un radio de 9, calcula el número de flechas que hay que recoger al bosque. 

# -------------
# Programa
import math

# Variables
num_flechazo=0
flechazo=[]
distancias=[]
quad=[0,0,0,0]
min_pos=0
min_num=0
max_pos=0
max_num=0
i=0

# Funciones
def dist_centro(x1,y1,x2,y2): # Función para calcular distancia entre 2 puntos
    return(math.sqrt((x2-x1)**2+(y2-y1)**2))

def pos_min_max(lista):
    min_pos=0
    min_num=0
    max_pos=0
    max_num=0
    i=0

    min_num=lista[0]
    for x in lista:
        if x < min_num:
            min_num=x
            min_pos=i
        if x > max_num:
            max_num=x
            max_pos=i
        i+=1
    return(min_pos,min_num,max_pos,max_num)

def outside_radius(lista,radio):
    num=0
    for x in lista:
        if x > radio:
            num+=1
    return(num)

# Código general - MAIN
for x in puntos:
    # 1 ¿Acierta Robin Hood? <-- Realizo el cálculo
    y=puntos.count(x)
    if y>1 and x not in flechazo:
        num_flechazo+=1
        flechazo.append(x)
    # 2 Flechazos por cuadrante <-- Realizo el cálculo
    if x[0]>=0 and x[1]>=0:
        quad[0]+=1
    elif x[0]<0 and x[1]>=0:
        quad[1]+=1
    elif x[0]<0 and x[1]<0:
        quad[2]+=1
    elif x[0]>=0 and x[1]<0:
        quad[3]+=1
    # 3 Punto más cercano al centro <-- Realizo el cálculo
    distancias.append(dist_centro(x[0],x[1],0,0)) # Almaceno en una lista las distancias al centro

    # 4 Diana radio  9. Número de flechas a recoger <-- Realizo el cálculo

    i+=1

# 1 ¿Acierta Robin Hood? <-- Muestro el resultado
print(f"Robin Hood tuvo doble flechazo {num_flechazo} veces en {flechazo}")

# 2 Flechazos por cuadrante <-- Muestro el resultado
i=1
for y in quad:
    print(f"Flechazos en cuadrante {i}: {y}")
    i+=1

# 3 Punto más cercano al centro <-- Muestro el resultado
min_pos,min_num,max_pos,max_num = pos_min_max(distancias)
print(f"El punto más cercano al centro es {puntos[min_pos]} y su distancia es {min_num}")

# 4 Diana radio  9. Número de flechas a recoger <-- Muestro el resultado
i = outside_radius(distancias,9)
print(f"Las flechas que no dieron en diana fueron: {i}")
```
